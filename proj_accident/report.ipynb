```python
import matplotlib.pyplot as plt
import numpy as np
import pylab
from clickhouse_driver import Client
client = Client('localhost')
```

```python
time = np.array(client.execute('SELECT time FROM parameter.param ORDER BY time')).flatten()

pressure = np.array(client.execute('SELECT pressure FROM parameter.param ORDER BY time')).flatten()

reactivity = np.array(client.execute('SELECT reactivity FROM parameter.param ORDER BY time')).flatten()

prz_water_level = np.array(client.execute('SELECT prz_water_level FROM parameter.param ORDER BY time')).flatten()

secondary_side_pressure = np.array(client.execute('SELECT secondary_side_pressure FROM parameter.param ORDER BY time')).flatten()

flow_rate_loop_1 = np.array(client.execute('SELECT flow_rate_loop_1 FROM parameter.param ORDER BY time')).flatten()
flow_rate_loop_2 = np.array(client.execute('SELECT flow_rate_loop_2 FROM parameter.param ORDER BY time')).flatten()
flow_rate_loop_3 = np.array(client.execute('SELECT flow_rate_loop_3 FROM parameter.param ORDER BY time')).flatten()

coolant_mass_loop_1 = np.array(client.execute('SELECT coolant_mass_loop_1 FROM parameter.param ORDER BY time')).flatten()
coolant_mass_loop_2 = np.array(client.execute('SELECT coolant_mass_loop_2 FROM parameter.param ORDER BY time')).flatten()
coolant_mass_loop_3 = np.array(client.execute('SELECT coolant_mass_loop_3 FROM parameter.param ORDER BY time')).flatten()

cladding_temperature_1 = np.array(client.execute('SELECT cladding_temperature_1 FROM parameter.param ORDER BY time')).flatten()
cladding_temperature_2 = np.array(client.execute('SELECT cladding_temperature_2 FROM parameter.param ORDER BY time')).flatten()
cladding_temperature_3 = np.array(client.execute('SELECT cladding_temperature_3 FROM parameter.param ORDER BY time')).flatten()

fuel_temperature_1 = np.array(client.execute('SELECT fuel_temperature_1 FROM parameter.param ORDER BY time')).flatten()
fuel_temperature_2 = np.array(client.execute('SELECT fuel_temperature_2 FROM parameter.param ORDER BY time')).flatten()
fuel_temperature_3 = np.array(client.execute('SELECT fuel_temperature_3 FROM parameter.param ORDER BY time')).flatten()

water_level_1 = np.array(client.execute('SELECT water_level_1 FROM parameter.param ORDER BY time')).flatten()
water_level_2 = np.array(client.execute('SELECT water_level_2 FROM parameter.param ORDER BY time')).flatten()
water_level_3 = np.array(client.execute('SELECT water_level_3 FROM parameter.param ORDER BY time')).flatten()

pump_velocity_loop_1 = np.array(client.execute('SELECT pump_velocity_loop_1 FROM parameter.param ORDER BY time')).flatten()
pump_velocity_loop_2 = np.array(client.execute('SELECT pump_velocity_loop_2 FROM parameter.param ORDER BY time')).flatten()
pump_velocity_loop_3 = np.array(client.execute('SELECT pump_velocity_loop_3 FROM parameter.param ORDER BY time')).flatten()
```

# Динамика аварийного процесса #

## 1. Давление в АЗ ##

```python
plt.figure(figsize=(15, 5), dpi= 80, facecolor='w', edgecolor='k')
plt.xlabel("Time, sec.", fontsize=16)
plt.ylabel("Pressure, Pa", fontsize=16)
plt.grid()

plt.xlim(left=0)
plt.ylim(bottom=0)
pylab.xticks(range(0, 6100, 250))
pylab.yticks(range(0, 14000000, 1000000))

plt.plot(time, pressure)
plt.show()
```

## 2. Реактивность ##

```python
plt.figure(figsize=(15, 5), dpi= 80, facecolor='w', edgecolor='k')
plt.xlabel("Time, sec.", fontsize=16)
plt.ylabel("Reactivity, $.", fontsize=16)
plt.grid()   

plt.xlim(left=0)
pylab.xticks(range(0, 6100, 250))
pylab.yticks(range(0, -31, -2))

plt.plot(time, reactivity)
plt.show()
```

## 3. Уровень воды в КД

```python
plt.figure(figsize=(15, 5), dpi= 80, facecolor='w', edgecolor='k')
plt.xlabel("Time, sec.", fontsize=16)
plt.ylabel("Level, m.", fontsize=16)
plt.grid()

plt.xlim(left=0)
pylab.xticks(range(0, 6100, 250))
pylab.yticks(range(0, 12, 1))


plt.plot(time, prz_water_level)
plt.show()
```

## 4. Давление во 2-ом контуре (ПГ)

```python
plt.figure(figsize=(15, 5), dpi= 80, facecolor='w', edgecolor='k')
plt.xlabel("Time, sec.", fontsize=16)
plt.ylabel("Pressure, Pa.", fontsize=16)
plt.grid()

plt.xlim(left=0)
pylab.xticks(range(0, 6100, 250))
pylab.yticks(range(0, 52000000, 100000))

plt.plot(time, secondary_side_pressure)
plt.show()
```

## 5. Расход теплоносителя в моделируемых петлях

```python
plt.figure(figsize=(15, 5), dpi= 80, facecolor='w', edgecolor='k')
plt.xlabel("Time, sec.", fontsize=16)
plt.ylabel("Flow rate, kg/s.", fontsize=16)
plt.grid()  

plt.xlim(left=0)
pylab.xticks(range(0, 6100, 250))
pylab.yticks(range(0, 4700, 500))

plt.plot(time, flow_rate_loop_1, label = 'Loop 1')
plt.plot(time, flow_rate_loop_2, label = 'Loop 2')
plt.plot(time, flow_rate_loop_3, label = 'Loop 3')
plt.legend()

plt.show()
```
